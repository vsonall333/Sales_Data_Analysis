```python
#ploting a bar chart for gender and its count
import seaborn as sns
ax = sns.countplot(x = 'Gender', data = sale_dataset, hue = "Gender")
for bars in ax.containers:
    ax.bar_label(bars)
```

#Age

```python
ax = sns.countplot(x='Age Group', data = sale_dataset, hue = 'Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Total amount vs age group

sale_age = sale_dataset.groupby(["Age Group"], as_index = False)["Amount"].sum().sort_values(by = "Amount", ascending = False)

# Define a list of colors (one for each bar)
colors = sns.color_palette("hsv", len(sale_age))

sns.barplot(x = "Age Group", y = "Amount", data = sale_age, palette=colors)
```

```python
# Total numbers of orders from top 10 states
sales_states = sale_dataset.groupby(['State'], as_index = False)["Orders"].sum().sort_values(by = "Orders", ascending = False). head(10)

# Define a list of colors (one for each bar)
colors = sns.color_palette("hsv", len(sales_states))

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_states, x = 'State', y = "Orders", palette=colors)
```

```python
# Total amount/sales from top 10 states

sales_states = sale_dataset.groupby(["State"], as_index = False)["Amount"].sum().sort_values(by = "Amount", ascending = False).head(10)

# Define a list of colors (one for each bar)
colors = sns.color_palette("hsv", len(sales_states))

sns.set(rc={"figure.figsize":(15, 6)})
sns.barplot(data = sales_states, x = "State", y = "Amount", palette=colors)
```

```python
ax = sns.countplot(data = sale_dataset, x = "Marital_Status", palette=colors)

sns.set(rc={"figure.figsize": (7,5)})

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_states = sale_dataset.groupby(["Marital_Status", "Gender"], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)

sns.set(rc = {'figure.figsize': (6,4)})

sns.barplot(data = sales_states, x = "Marital_Status", y = "Amount", hue = "Gender")
```

# Occupation

```python
# Checking the number of profession
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = sale_dataset, x = 'Occupation', palette=colors)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#Analyzing the purchasing power of various occupations.

sales_state = sale_dataset.groupby(["Occupation"], as_index = False)["Amount"].sum().sort_values(by="Amount", ascending = False)
sns.set(rc={'figure.figsize': (20, 5)})
sns.barplot(data = sales_state, x = "Occupation", y = "Amount", palette = colors)
```

From the above graphs we interpret that people belong from IT, Healthcare, and Aviation sector spend more

## Product Category

```python
#Finding the number of product sold

sns.set(rc = {'figure.figsize':(25,6)})

ax = sns.countplot(data = sale_dataset, x = 'Product_Category', palette=colors)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Finding the highest amount spend in which product

sale_product = sale_dataset.groupby(['Product_Category'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False).head(10)

sns.set(rc = {"figure.figsize":(25,5)})
sns.barplot(data = sale_product, x = "Product_Category", y = 'Amount', palette = colors)
```
